####Usando librería pandas para reordenar la información

Pondremos los datos de los profesores en una tabla ocupando la libreria Pandas. Para esto primero debemos definir que datos son los que vamos a ocupar.

Tenemos los campos de : Nombre, Rol, Email y Anexo. Es importante destacar que hay profesores que tienen 2 roles, por lo que el campo "Rol" se transforma en "Rol 1" y "Rol 2"

````python
import pandas as pd
````

````python
headers = ["Nombre", "Rol 1", "Rol 2", "Email", "Anexo"]
````

````python
#Modifiquemos el código de antes para guardar los atributos
nombres = [] #lista vacia en donde guardaremos los nombres
rol1 = []
rol2 = []
emails = []
anexos = []

for bar in soup.find_all('div', attrs={'class': 'ficha-persona'}): #me encuentra todos
  nombres.append(bar.find('div', attrs={'class': 'nombre'}).text)
  cargo = bar.find('div', attrs={'class': 'cargo'}).text
  #Si un profesor tiene 2 cargos, el simbolo '|' los separa.
  doble = cargo.find('|') #buscamos el simbolo '|' en el string
  
  
  if(doble!=-1): #si posee dos roles
    rol1.append(cargo[0:doble-1])
    rol2.append(cargo[doble+1:])
  else: #si no posee 2 roles
    rol1.append(cargo)
    rol2.append(float("NAN"))

  emails.append(bar.find('a', attrs={'class': 'mail'}).text)
  
  if(bar.find('div', attrs={'class': 'fono'})):
    anexos.append(bar.find('div', attrs={'class': 'fono'}).text)
  else:
    anexos.append(float("NAN"))
academicos= [nombres,rol1,rol2,emails,anexos]
````

````python
new_table = pd.DataFrame()
for i in range(len(headers)):
    new_table[headers[i]] = academicos[i]

new_table
````

###Ejemplo avanzado 


Para este ejemplo mostraré cómo extraer información sobre las ofertas de juegos disponibles en la plataforma de [Steam](https://store.steampowered.com/search/?specials=1). Para lograrlo, aplicaré uso de las herramientas de Python que vimos anteriormente, en especial, **Beautifulsoup** para obtener los datos del html y **Pandas** para guardar los datos importantes en una tabla.

En este caso, los datos que nos interesan son los siguientes:

- Nombre del juego
- Fecha de lanzamiento
- Precio original
- Precio en oferta
- Review de los usuarios (Positivas, negativas, neutro)
- Plataformas en la que se puede jugar (Windows, Linux y/o Mac)
- Etiqueta del juego

Además, haré uso de un archivo llamado "tags.csv" el cual contiene todas las etiquetas que puede tener un juego de steam con su código para hacer la búsqueda a través de un **requests**. El primer paso es  leer este archivo (utilizando pandas) y hacer un request con cada código para obtener los datos anteriormente mencionados.

* Además, se verá cómo realizar filtrado y manipulación de información.
* Todos los datos obtenidos deberán ser guardados en un archivo ".csv" 

````python
import requests
import bs4
import pandas as pd
````

````python
TAGS_DF = pd.read_csv('tags.csv', index_col=0)
tags = TAGS_DF["Etiqueta"]
cods = TAGS_DF["Código"]

nombres   = []
fecha     = []
p_orig    = []
p_oferta  = []
reviews   = []
plats     = []
etiquetas = []

for i in range(len(tags)):
    tag = tags[i]
    cod = cods[i]
    r = requests.get(url = "https://store.steampowered.com/search/", params = { "specials": 1, "tags": cod , "cc": "cl" })
    soup = bs4.BeautifulSoup(r.text, "lxml")
    
    for game in soup.find_all('div', 'responsive_search_name_combined'):
        # Nombres
        nombres.append( game.find('span', 'title').text )
        # Fecha
        f = game.find('div', 'col search_released responsive_secondrow')
        if f: fecha.append( f.text )
        else: f.append( "" )
            
        # Precios
        p = game.find('div', attrs={'search_price'})
        if p:
            p = p.text.replace('CLP$', '').replace('.','').split()
            if len(p) == 1:
                if p[0] == 'Free': p_or, p_of = float('nan'), float('nan')
                else: p_or, p_of = int(p[0]) if('Free' not in p[0]) else int(p[0].replace('Free', '')) , float('nan')
            elif len(p) == 2: p_or, p_of = int(p[0]), int(p[1])
            else: p_or, p_of = float('nan'), float('nan')
            p_orig.append(p_or)
            p_oferta.append(p_of)
        else:
            p_orig.append( float('nan') )
            p_oferta.append( float('nan') )
            
        # Review
        r = game.find('div', attrs={'search_reviewscore'})
        if r.span:
            reviews.append(r.span["data-tooltip-html"].split('<br>')[0])
        else: reviews.append( "" )
            
        # Plataformas
        pl = game.find_all('span', attrs={'platform_img'})
        aux_p = []
        if pl:
            for p in pl:
                p = p["class"]
                p.remove('platform_img')
                if not p: continue
                if 'win' in p: aux_p.append('Windows')
                if 'mac' in p: aux_p.append('Mac')
                if 'linux' in p: aux_p.append('Linux')
            aux_p = ", ".join(aux_p)
        else: aux_p = ""
        plats.append(aux_p)
        
        #Etiquetas
        etiquetas.append(tag)
        
steam = pd.DataFrame({ 'Nombre del juego':nombres, 'Fecha de lanzamiento':fecha, "Precio original": p_orig,
                      "Precio oferta": p_oferta, "Review": reviews, "Plataformas": plats, "Etiqueta": etiquetas })
    
display(steam.head())
````

**En toda extracción de texto es importante tomar decisiones para tratar los datos, en mi caso tomé las siguientes para la representación de los datos.**

| Columna | Tipo de dato esperado | Datos posibles / intervalos | Datos faltantes |

Sidenotes:
* Los **datos semiestructurados** se refieren a cualquier información que utilice un esquema de autodescripción, como XML o JSON.
* Los **datos estructurados** incluyen nombres, fechas, direcciones, números de tarjetas de crédito, información bursátil, geolocalización y más.
